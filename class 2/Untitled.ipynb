```python
def KNN(X,Y,point):
    ans = []
    for i in range(len(X)):
        dis = np.sqrt(np.sum((X[i,:] - point)**2))
        ans.append([dis,Y[i]])
    ans.sort()
    counter0 = 0
    counter1 = 0
    for i in range(5):
        if(ans[i][1] == 0):
            counter0 += 1
        else:
            counter1 += 1
            
    if(counter0 > counter1):
        return 0
    else:
        return 1
        
    
```

```python
import numpy as np
point  = np.array([1,1])
answer  = KNN(X,Y,point)

plt.scatter(X[:,0],X[:,1],c = Y)
plt.scatter(point[0],point[1],c = "red")
plt.show()
print(answer)

```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

# y = x


def gradient_decent(x,y):
    m = 0 
    c = 0
    learning_rate = 0.001
    
    for i in range(10000):
        m = m + learning_rate * np.sum((y-m*x-c)*x)
        c = c + learning_rate * np.sum(y-m*x-c)
        plt.plot(x,m*x+c)
        
    return m,c

m,c = gradient_decent(x,y)
        
plt.show()
```
